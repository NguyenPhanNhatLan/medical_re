```python
import json
from pathlib import Path

# ====== SỬA TÊN 3 FILE Ở ĐÂY ======
INPUT_FILES = [
    "rel_Quynh.json",
    "rel_Thu.json",
    "rel_Tien.json",
]

OUT_FULL_JSON = "full.json"


def load_json(fp: str):
    p = Path(fp)
    if not p.exists():
        raise FileNotFoundError(f"Không thấy file: {p.resolve()}")
    with p.open("r", encoding="utf-8") as f:
        return json.load(f)


def main():
    merged = []

    for fp in INPUT_FILES:
        data = load_json(fp)

        # Trường hợp phổ biến: mỗi file là list tasks
        if isinstance(data, list):
            merged.extend(data)
        else:
            raise ValueError(
                f"File {fp} không phải top-level LIST. "
                f"Nếu file của bạn là dict (object), nói mình để mình chỉnh kiểu merge phù hợp."
            )

    with open(OUT_FULL_JSON, "w", encoding="utf-8") as f:
        json.dump(merged, f, ensure_ascii=False, indent=2)

    print(f"✅ Merged {len(INPUT_FILES)} files -> {OUT_FULL_JSON}")
    print(f"✅ Total tasks in full: {len(merged)}")


if __name__ == "__main__":
    main()
```
